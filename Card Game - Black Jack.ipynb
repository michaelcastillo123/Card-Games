```python
import random
suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
value = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six': 6, 'Seven':7, 'Eight':8, 'Nine':9, 'Ten':10, 'Jack':11, 'Queen':12,'King':13, 'Ace':1}
```

```python
class Card():
    
    def __init__(self,suit,rank):
        self.suit = suit
        self.rank = rank
        self.value = value[rank]
    
    def __str__(self):
        return f'{self.value} of {self.suit}'
```

```python
class Player():
    
    def __init__(self,name):
        self.name = name
        self.hand_up = []
        self.hand_down = []
        self.value_up = []
        self.value_down =[]
            
    def add_card_up(self,new_card):
        return self.hand_up.append(new_card)
    
    def add_card_down(self,new_card):
        return self.hand_down.append(new_card)
```

```python
class Deck():
    
    def __init__(self):
        self.deck = []
        for s in suits:
            for r in ranks:
                created_card = Card(s,r)
                self.deck.append(created_card)
        
    def shuffle(self):
        random.shuffle(self.deck)
        
    def deal_one(self):
        return self.deck.pop(0)
```

```python
p_name = input('What is your name?\n')

#this while loop is to replay
while True:
    
    #Set up Players
    dealer = Player('Dealer')
    p1 = Player(p_name)
    
    #prepare deck
    cards = Deck()
    cards.shuffle()

    #distribute two cards per player
    for i in range(1):
        dealer.add_card_up(cards.deal_one())
        dealer.add_card_down(cards.deal_one())
        p1.add_card_up(cards.deal_one())
        p1.add_card_down(cards.deal_one())

    #beginning sum of each players hand faced up
    sum_d = 0
    sum_p1 = 0

    #Show cards faced up
    list_d = [dealer.hand_up[0].value]
    list_p1 = [p1.hand_up[0].value]

    d_hit_stay = 0
    p1_hit_stay = 0

    while True:

    #establish the sum of faced up and down separately  
        black_jack = 21
        sum_d = sum(list_d)
        sum_p1 = sum(list_p1)

        sum_d_all = dealer.hand_down[0].value
        sum_p1_all = p1.hand_down[0].value

        sum_d_all = sum_d_all + sum_d
        sum_p1_all = sum_p1_all + sum_p1

        print('--------------------------------------------')
        print(f'\n{dealer.name} has {(list_d)} faced up ')
        #print(f'and {dealer.hand_down[0].value} down')
        print(f'{p1.name} has {(list_p1)} faced up')
        print(f'and {p1.hand_down[0].value} down\n')
        #print(f'***p1 has {sum_p1_all}')
        #print(f'***dealer has {sum_d_all}')

    #check for winner
        if sum_d > 21 or sum_p1 > 21:
            if sum_d_all > black_jack:
                print(f'Game over! {dealer.name} bust')
            elif sum_p1_all > black_jack:
                print(f'Game over! {p1.name} bust') 
            print(f'Dealer had {sum_d_all} and {p1.name} had {sum_p1_all}')
            break
            False

        if d_hit_stay > 0 and p1_hit_stay > 0:
            if sum_d_all == black_jack:
                print(f'{dealer.name} won with a blackjack!')
            elif sum_p1_all == black_jack:
                print(f'{p1.name} won with a blackjack!')
            elif sum_d_all > black_jack:
                print(f'Game over! {dealer.name} bust')
            elif sum_p1_all > black_jack:
                print(f'Game over! {p1.name} bust')   
            elif sum_d_all > sum_p1_all:
                print(f'{dealer.name} won!')
            elif sum_d_all < sum_p1_all:
                print(f'{p1.name} won!')
            print(f'Dealer had {sum_d_all} and {p1.name} had {sum_p1_all}')
            break
            False

    #check for bust each round with hand faced up
        if sum_d > black_jack:
            print(f'Game over! {dealer.name} bust')
            break
        if sum_p1 > black_jack:
            print(f'Game over! {p1.name} bust')
            break

    #ask for hit or stay
        d_hit_stay = 0
        if sum_d_all < 17:
            dealer.add_card_up(cards.deal_one())
            print(f'{dealer.name} drew {dealer.hand_up[-1]}')
            list_d.append(dealer.hand_up[-1].value)
        else:
            d_hit_stay += 1

        while True:
            if p1_hit_stay > 0:
                break
                False
            hit_stay = input('Would you like to stay or hit?\n').title()
            if hit_stay == 'Hit' or hit_stay == 'H':
                p1.add_card_up(cards.deal_one())
                print(f'{p1.name} drew {p1.hand_up[-1]}')
                list_p1.append(p1.hand_up[-1].value)
                break
            elif hit_stay == 'Stay' or hit_stay == 'S':
                p1_hit_stay = 1
                break
                
    start_again = input('Would you like to play again y or n?\n').title()
    if start_again == 'N' or start_again == 'No':
        break
    else:
        continue
            

            
```
